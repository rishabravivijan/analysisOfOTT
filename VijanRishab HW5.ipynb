```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataframe= pd.read_csv("MoviesOnStreamingPlatforms_updated.csv")
```

```python
dataframe.shape
```

## Data Analysis

```python
dataframe.drop(['Unnamed: 0','ID','Type'],axis =1,inplace=True)
```

```python
dataframe
```

```python
dataframe.isnull().sum()
```

```python
print(dataframe.median())
```

```python
median_runtime = dataframe["Runtime"].median()
```

```python
# dataframe = dataframe.fillna(value=dataframe['Runtime'].median())
dataframe["Runtime"].fillna(median_runtime, inplace = True)
# data_new["Runtime"].fillna(median_runtime, inplace = True)
```

```python
dataframe.isnull().sum()
```

## Heat Map

```python
correaltion = dataframe.corr()
sns.heatmap(correaltion, annot=True, square=True)
plt.yticks(rotation=2)
plt.show()
```
